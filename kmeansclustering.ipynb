```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.array([
    [1, 2], [2, 1], [3, 4], [5, 7], [3, 5], [8, 7], [6, 9], [7, 5], [8, 6],
    [15, 18], [16, 20], [14, 17], [13, 16], [18, 22], [19, 21],
    [25, 5], [24, 6], [23, 5], [26, 7], [22, 4],
])

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
clusters = kmeans.labels_
centroids = kmeans.cluster_centers_
colors = ['red', 'blue','violet']
for i in range(3):
    points = data[clusters == i]
    plt.scatter(points[:, 0], points[:, 1], c=colors[i], label=f'Cluster {i+1}')
plt.scatter(centroids[:, 0], centroids[:, 1], c='green', marker='X', s=100, label='Centroids')
plt.legend()
plt.show()
```
