```python
import pandas as pd
import matplotlib.pyplot as plt
data1={"semester":["1","2","3","4"],
       "percentage":[89.9,91.2,90,95.4]}
df=pd.DataFrame(data1)
print(df)
```

```python
plt.plot(df["semester"],df["percentage"],color="red",marker="o")
plt.title("Student results chart")
plt.xlabel("Semester")
plt.ylabel("Percentage scored")
plt.show()
```
